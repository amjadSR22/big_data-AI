```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
gender = pd.read_csv("name_gender_dataset.csv")
gender
```

```python
gender.info()
```

```python
# line plot # matplotlib
plt.figure(figsize=(10,5))# Appropriate figure size
plt.title('Relation between the count and probability') # title
plt.xlabel('Probability') # X label
plt.ylabel('Count') # y label
plt.plot(gender['Probability'], gender['Count']);
# Explain 
# the graph it's show direct relationship between Count and probability 
```

```python
gender_index = gender.iloc[0:100]
# the data took a lot of time to process that's why I took a small sample
```

```python
#Bar Chart # matplotlib
plt.figure(figsize=(10,5)) # Size
plt.title('number of babies') # title
plt.xlabel('gender') # x label
plt.ylabel('count') # y label
plt.bar(gender_index['Gender'], gender_index['Count']);
# explain 
# the number of male from babies is more than female
```

```python
#Scatter Plot # matplotlib
plt.figure(figsize=(10,5))
plt.scatter( gender['Probability'] , gender['Count']);
plt.title('The Realtion between Count and probability'); # title
plt.xlabel('Probability'); # xlabel
plt.ylabel('Count'); # ylabel
#explain 
# there a direct relationship between them
# also is seem like there something weird cause there no points at the middle
```

```python
#number of female
num_f = (gender['Gender'] == 'F').sum()
num_f
```

```python
#number of male
num_m = (gender['Gender'] == 'M').sum()
num_m
```

```python
# pie chart # matplotlib  # seaborn help for customize the graph
plt.figure(figsize= (10,20)) # SIZE
plt.title('number of male and female') # title
num_gen = [num_f , num_m] # number of man and female
label = ['Female' , 'Male'] 
plt.pie(num_gen , labels= label);
plt.legend(); # legend
sns.set()
sns.color_palette('pastel');
# explain
# the number of female in all sample is greater than male
```

```python
# Histogram plot # matplotlib
plt.figure(figsize=(10,5));
plt.hist(gender_index['Gender']);
plt.title("Number of male and female") # title
plt.xlabel('Gender'); # xlabel
plt.ylabel('Count'); # ylabel
#explain
#number of male more than female 
```
